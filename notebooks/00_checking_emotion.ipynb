```python
import pandas as pd
import os

# Path dan output file
input_path = '../data/raw/datatrain.csv'
output_dir = '../data/processed'
output_path = os.path.join(output_dir, 'train_clean.csv')

# Load the CSV file
df_train = pd.read_csv(input_path)

# Dictionary untuk mengoreksi dan menyamakan nilai emosi
emotion_map = {
    'Terkejut': 'Surprise',
    'Trkejut': 'Surprise',
    'Terkjut': 'Surprise',
    'Trekejut': 'Surprise',
    'Kaget': 'Surprise',
    'kaget': 'Surprise',
    'terkejut': 'Surprise',
    'Marah': 'Anger',
    'Marh': 'Anger',
    'Bangga': 'Proud',
    'Percaya': 'Trust',
    'Loyalty': 'Trust',
    'Faith': 'Trust',
    'Sad': 'Sadness'
}

# mengganti nilai yang tidak konsisten di kolom 'emotion'
df_train['emotion'] = df_train['emotion'].replace(emotion_map)

# menghapus baris yang memiliki emosi 'Love' karena tidak termasuk dalam 8 kategori BDC
df_train = df_train[df_train['emotion'] != 'Love']

# memastikan folder output ada, jika tidak, buat
os.makedirs(output_dir, exist_ok=True)

# menyimpan DataFrame yang sudah dibersihkan ke file CSV
df_train.to_csv(output_path, index=False)

print(f"\nDataFrame yang sudah dibersihkan sudah disimpan di '{output_path}'")
print(f"Sekarang total baris: {len(df_train)} (satu baris 'Love' telah dihapus).")
print("\nNilai unik setelah perbaikan:")
print(df_train['emotion'].unique())
```

```python
import pandas as pd

# Path ke file output
output_path = '../data/processed/train_clean.csv'

try:
    # Load the cleaned CSV file
    df_clean = pd.read_csv(output_path)

    print("--- Mengecek Keunikan Kolom 'emotion' ---")
    # Hitung nilai unik dan frekuensinya di kolom 'emotion'
    emotion_unique_counts = df_clean['emotion'].value_counts()
    print("Nilai unik dan jumlahnya di kolom 'emotion':")
    print(emotion_unique_counts)

    # Periksa apakah semua nilai emosi sudah sesuai dengan daftar yang diinginkan
    print("\nNilai unik yang tersisa di kolom 'emotion' setelah normalisasi:")
    print(df_clean['emotion'].unique())

    print("\n--- Mengecek Keunikan Kolom 'id' ---")
    # Periksa apakah ada ID duplikat
    total_rows = len(df_clean)
    unique_ids = df_clean['id'].nunique()
    if total_rows == unique_ids:
        print(f"Jumlah ID unik ({unique_ids}) sama dengan jumlah baris ({total_rows}). Semua ID unik.")
    else:
        print(f"ADA ID duplikat! Jumlah ID unik ({unique_ids}) tidak sama dengan jumlah baris ({total_rows}).")

    print("\n--- Mengecek Informasi DataFrame ---")
    print("Informasi umum tentang DataFrame:")
    print(df_clean.info())
    
except FileNotFoundError:
    print(f"Error: File '{output_path}' tidak ditemukan.")
    print("Pastikan file 'train_clean.csv' sudah berhasil dibuat di folder 'data/processed' terlebih dahulu.")
```
