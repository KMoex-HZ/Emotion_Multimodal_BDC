# Download Video

```python
import pandas as pd
import os
import yt_dlp

def download_videos(csv_path, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    df = pd.read_csv(csv_path)
    
    failed_downloads = []

    for _, row in df.iterrows():
        vid, url = row['id'], row['video']
        outpath = os.path.join(out_dir, f"{vid}.mp4")
        if os.path.exists(outpath):
            print(f"✅ Video {vid} sudah ada, lanjut.")
            continue
        try:
            ydl_opts = {
                'outtmpl': outpath,
                'format': 'mp4/best',
                'quiet': True
            }
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
                print(f"✅ Berhasil mengunduh video {vid}")
        except Exception as e:
            print(f"❌ Gagal mengunduh video {vid}: {e}")
            failed_downloads.append({'id': vid, 'url': url})

    if failed_downloads:
        print("\n---")
        print("🚨 Berikut daftar video yang gagal diunduh:")
        for item in failed_downloads:
            print(f"- ID: {item['id']}, URL: {item['url']}")
    else:
        print("\n---")
        print("🎉 Semua video berhasil diunduh tanpa kendala.")

download_videos("../data/datatrain.csv", "../data/videos/train")
download_videos("../data/datatest.csv", "../data/videos/test")
```

# Penanganan video gagal download

```python
failed_train = download_videos("../data/datatrain.csv", "../data/videos/train")
failed_test  = download_videos("../data/datatest.csv", "../data/videos/test")

if failed_train:
    pd.DataFrame(failed_train).to_csv("../data/failed_train.csv", index=False)
    print("🚨 Daftar video train gagal disimpan di ../data/failed_train.csv")

if failed_test:
    pd.DataFrame(failed_test).to_csv("../data/failed_test.csv", index=False)
    print("🚨 Daftar video test gagal disimpan di ../data/failed_test.csv")

def retry_failed(failed_csv, out_dir):
    if not os.path.exists(failed_csv):
        return
    df = pd.read_csv(failed_csv)
    if df.empty:
        return
    print(f"🔄 Mencoba ulang {len(df)} video dari {failed_csv}")
    failed_downloads = []
    for _, row in df.iterrows():
        vid, url = row['id'], row['url']
        outpath = os.path.join(out_dir, f"{vid}.mp4")
        try:
            ydl_opts = {
                'outtmpl': outpath,
                'format': 'mp4/best',
                'quiet': True
            }
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
                print(f"✅ Retry berhasil: {vid}")
        except Exception as e:
            print(f"❌ Retry gagal lagi {vid}: {e}")
            failed_downloads.append({'id': vid, 'url': url})
    if failed_downloads:
        pd.DataFrame(failed_downloads).to_csv(failed_csv, index=False)
        print(f"🚨 Masih ada gagal, tersimpan lagi di {failed_csv}")
    else:
        print("🎉 Semua retry berhasil!")

retry_failed("../data/failed_train.csv", "../data/videos/train")
retry_failed("../data/failed_test.csv", "../data/videos/test")
```

# Menghitung missing data

```python
import pandas as pd
import os

def cek_missing(csv_path, video_dir):
    df = pd.read_csv(csv_path)
    total = len(df)
    downloaded = sum(os.path.exists(os.path.join(video_dir, f"{vid}.mp4")) for vid in df['id'])
    missing = total - downloaded

    print(f"📊 Dari {total} video di {os.path.basename(csv_path)}:")
    print(f"✅ Berhasil diunduh: {downloaded}")
    print(f"❌ Gagal / Missing: {missing}")

    missing_ids = [vid for vid in df['id'] if not os.path.exists(os.path.join(video_dir, f"{vid}.mp4"))]
    return missing_ids

missing_train = cek_missing("../data/datatrain.csv", "../data/videos/train")
missing_test  = cek_missing("../data/datatest.csv", "../data/videos/test")
```
