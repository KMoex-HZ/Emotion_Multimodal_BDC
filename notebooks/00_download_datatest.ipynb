```python
import pandas as pd
import os
import yt_dlp
import shutil

# Path untuk file CSV dan folder output
csv_path = '../data/raw/datatest_revisi.csv'
output_dir = '../data/video/test'

def download_videos(csv_path, out_dir):
    """Mengunduh video dari URL yang ada di file CSV."""
    
    # Pastikan folder output ada
    os.makedirs(out_dir, exist_ok=True)
    
    try:
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f"‚ùå Error: File '{csv_path}' tidak ditemukan. Pastikan file berada di folder yang benar.")
        return

    failed_downloads = []

    # Loop untuk setiap baris di DataFrame
    for _, row in df.iterrows():
        vid, url = row['id'], row['video']
        outpath_temp = os.path.join(out_dir, f"{vid}_temp")
        outpath = os.path.join(out_dir, f"{vid}.mp4")

        # Cek apakah video sudah diunduh sebelumnya
        if os.path.exists(outpath):
            print(f"‚úÖ Video {vid} sudah ada, lanjut.")
            continue

        try:
            ydl_opts = {
                'outtmpl': outpath_temp,
                'format': 'mp4/best',
                'quiet': True,
                'retries': 3,
                'no_warnings': True,
                'nocheckcertificate': True
            }
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
            
            # Pindahkan file dari temp ke final
            shutil.move(outpath_temp, outpath)
            
            print(f"‚úÖ Berhasil mengunduh video {vid}")
        except Exception as e:
            print(f"‚ùå Gagal mengunduh video {vid} dari {url}. Error: {e}")
            failed_downloads.append({'id': vid, 'url': url})

    if failed_downloads:
        print("\n---")
        print("üö® Berikut daftar video yang gagal diunduh:")
        for item in failed_downloads:
            print(f"- ID: {item['id']}, URL: {item['url']}")
    else:
        print("\n---")
        print("üéâ Semua video berhasil diunduh tanpa kendala.")

# Panggil fungsi untuk mengunduh data test yang baru
download_videos(csv_path, output_dir)
```
